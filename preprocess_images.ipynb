```python
import os
from PIL import Image


path_to_cat='images/cat/archive/CAT_01'
x = os.listdir(path_to_cat)
size = (50,50)
new_path='images/cat/500cats/'
for index,image_path in enumerate(x):
    if index>500:
        break
    if image_path.endswith('.jpg'):
        print(image_path)
        im = Image.open(path_to_cat+'/'+image_path)
        im=im.resize(size)
        im.save(new_path+str(index)+'.jpg', "JPEG")



```

```python

path_to_dog='images/dog/archive(1)/images/Images'
x = os.listdir(path_to_dog)
size = (50,50)
new_path='images/dog/500dogs/'

for z,item in enumerate(x):
    dog_folder = os.listdir(path_to_dog+'/'+item)
    for index,image_path in enumerate(dog_folder):
        if index>5:
            break
        if image_path.endswith('.jpg'):
            print(image_path)
            im = Image.open(path_to_dog+'/'+item+'/'+image_path)
            im=im.resize(size)
            im.save(new_path+str(z)+str(index)+'.jpg', "JPEG")


```
